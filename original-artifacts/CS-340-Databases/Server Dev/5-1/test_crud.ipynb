```python
# Import the AnimalShelter class
from animal_shelter import AnimalShelter
```

```python
# Instantiate the class with aacuser credentials
shelter = AnimalShelter(username="aacuser", password="SNHU1234")
```

```python
# Test Create
test_data = {
    "animal_id": "A999999",
    "animal_type": "Dog",
    "name": "TestDog",
    "breed": "Labrador Retriever",
    "outcome_type": "Adoption",
    "age_upon_outcome": "1 year",
    "sex_upon_outcome": "Neutered Male"
}
create_result = shelter.create(test_data)
print(f"Create result: {create_result}")
```

```python
# Test Read
query = {"animal_type": "Dog"}
read_result = shelter.read(query)
print(f"Read result (first 2 documents):")
for doc in read_result[:2]:
    print(doc)
```

```python
# Test Update
update_query = {"animal_id": "A999999"}
update_data = {"name": "UpdatedTestDog"}
update_result = shelter.update(update_query, update_data)
print(f"Update result: {update_result} document(s) modified")
```

```python
# Test Delete
delete_query = {"animal_id": "A999999"}
delete_result = shelter.delete(delete_query)
print(f"Delete result: {delete_result} document(s) deleted")
```
